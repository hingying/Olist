Remove stopwords and punctuation and hyperlink

```python
def re_breakline(text_list):
    """
    Args:
    ----------
    text_list: list object with text content to be prepared [type: list]
    """
    
    # Applying regex
    return [re.sub('[\n\r]', ' ', r) for r in text_list]
```

```python
def re_hiperlinks(text_list):
    """
    Args:
    ----------
    text_list: list object with text content to be prepared [type: list]
    """
    
    # Applying regex
    pattern = 'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+'
    return [re.sub(pattern, ' link ', r) for r in text_list]
```

```python
new['review_comment_message']=new['review_comment_message'].apply(str)
```

```python
new['review_comment_message']=new['review_comment_message'].apply(lambda x:re.sub('[\n\r]', ' ', x) if x!=None else x)
```

```python
new['review_comment_message']=new['review_comment_message'].apply(lambda x:re.sub('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', 'link', x) if x!=None else x)
```

```python
import string
def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text
```

```python
new['review_comment_message']=new['review_comment_message'].apply(lambda x: remove_punctuations(x) if x!=None else x)
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
pat = r'\b(?:{})\b'.format('|'.join(stop))
new['tweet_without_stopwords'] = new['review_comment_message'].str.replace(pat, '')
new['tweet_without_stopwords'] = new['tweet_without_stopwords'].str.replace(r'\s+', ' ')
```
